# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending per student actually (\$645-675) underperformed compared to schools with smaller budgets (<\$585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
# Dependencies and Setup
import pandas as pd
import numpy as np
from pprint import pprint as pp

# File to Load (Remember to Change These)
school_data_to_load = "schools_complete.csv"
student_data_to_load = "students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
school_data_complete.head()
```

```python


schoolCount =len(school_data_complete["school_name"].unique())
studentCount = len(school_data_complete["Student ID"].unique())
budgetTotal = sum(school_data_complete["budget"].unique())
mathAverage = (sum(school_data_complete["math_score"])/studentCount)
readingAverage = (sum(school_data_complete["reading_score"])/studentCount)
overallPass = (mathAverage + readingAverage)/2


```

```python
# define what is a pass and failing score

bins = [0, 69.99999, 100]
group_labels = ["fail", "pass"]
pd.cut(school_data_complete["math_score"], bins, labels = group_labels).head()
pd.cut(school_data_complete["reading_score"], bins, labels = group_labels).head()
```

```python
# label each score as passing or failing
school_data_complete["Reading_Pass/Fail"] = pd.cut(school_data_complete["reading_score"], bins, labels = group_labels)
school_data_complete["Math_Pass/Fail"] = pd.cut(school_data_complete["math_score"], bins, labels = group_labels)

school_data_complete.head()
```

```python
math_pass = school_data_complete.loc[school_data_complete["Math_Pass/Fail"] == "pass", :]
reading_pass = school_data_complete.loc[school_data_complete["Reading_Pass/Fail"] == "pass", :]
```

```python
math_pass.head()
```

```python
reading_pass.head()
```

```python
mathPassCount = len(math_pass)
readPassCount = len(reading_pass)
print(mathPassCount)
print(readPassCount)
```

```python
mathPassPercent = (mathPassCount/studentCount)*100
readPassPercent = (readPassCount/studentCount)*100
print(mathPassPercent)
print(readPassPercent)
```

## Top Performing Schools (By Passing Rate)

```python
school_data_complete.head()
```

```python
school_financial = school_data_complete[['school_name', 'School ID', 'type', 'size', 'budget']].copy()
```

```python
school_financial.head()
```

```python
school_data = school_financial.drop_duplicates()
```

```python
schoolList = school_data.set_index('school_name')
schoolList.head()
```

```python
schools = schoolList.index.values.tolist()
```

```python
schools
```

```python
budgetData= {}
keys = schools
perStudent = (schoolList['budget']/schoolList['size'])
for i in keys:
    budgetData[i] = perStudent[i]
pp(budgetData)
```

```python
studentBudget = pd.DataFrame.from_dict(budgetData, orient='index', columns = ['Per Student Budget'])
studentBudget.index.name = 'school_name'
studentBudget.head()
```

```python
districtSummary1 = pd.merge(schoolList, studentBudget, on = 'school_name' )
districtSummary1
```

```python
school_results = school_data_complete[['school_name', 'School ID', 'reading_score', 'math_score', 'size', 'Reading_Pass/Fail', 'Math_Pass/Fail']].copy()
school_results.head()
```

```python
school_results.groupby(['school_name']).groups.keys()
len(school_results.groupby(['school_name']).groups['Huang High School'])
```

```python
aveReading = school_results.groupby('school_name')[['reading_score']].mean()
type(aveReading)
```

```python
aveMath = school_results.groupby('school_name')[['math_score']].mean()
type(aveMath)
```

```python
aveScores = school_results.groupby('school_name').agg({'math_score': 'mean',
                                          'reading_score': 'mean'})
# aveScores.columns = aveScores.columns.droplevel(level=O)
# aveScores.rename(columns={'school_name': 'School Name', 'math_score': 'Average Math Score', 'reading_score': 'Average Reading Score'})
aveScores
```

```python
readingPass = school_results[school_results['Reading_Pass/Fail'] == 'pass'].groupby('school_name')[['Reading_Pass/Fail']].count()


```

```python
mathPass = school_results[school_results['Math_Pass/Fail'] == 'pass'].groupby('school_name')[['Math_Pass/Fail']].count()
```

```python
studentPass = pd.merge(readingPass, mathPass, on = 'school_name' )
studentPass
```

```python
districtSummary2 = pd.merge(districtSummary1, studentPass, on = 'school_name' )
```

```python
districtSummary2['% Passing Math'] = np.where(districtSummary2['size'] < 0, 0, districtSummary2['Math_Pass/Fail']/districtSummary2['size'])
districtSummary2['% Passing Reading'] = np.where(districtSummary2['size'] < 0, 0, districtSummary2['Reading_Pass/Fail']/districtSummary2['size'])
```

```python
districtSummary2['Overall Pass %'] = (districtSummary2['% Passing Math'] + districtSummary2['% Passing Reading']) / 2
districtSummary2.head()
```

```python
districtSummary3 = pd.merge(districtSummary2, aveScores, on = 'school_name' )
districtSummary3
```

```python
districtSummary3.columns = ['School ID', 'School Type', 'Total Students', 'Total School Budget', 'Per Student Budget', 'Reading_Pass/Fail', 'Math_Pass/Fail', '% Passing Math',
                           '% Passing Reading', 'Overall Pass %' , 'Average Math Score', 'Average Reading Score']
districtSummary3
```

```python
SchoolTable = districtSummary3[['School Type', 'Total Students', 'Total School Budget', 'Per Student Budget', 
                                'Average Math Score', 'Average Reading Score', '% Passing Math', '% Passing Reading', 'Overall Pass %']]
SchoolTable.head()
```

```python
SchoolTableFormat = SchoolTable.style.format({'Total Students' : "{:,}", 
                                                  'Total School Budget' : "${:,.0f}",
                                                  'Per Student Budget' : "${:,.0f}",
                                                  'Average Math Score' : "{:.2f}",
                                                  'Average Reading Score' : "{:.2f}",
                                                  '% Passing Math' : "{:.2%}",
                                                  '% Passing Reading' : "{:.2%}",
                                                  'Overall Pass %' : "{:.2%}"})

SchoolTableFormat
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
SchoolTableBudget = SchoolTable.copy()
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
SchoolTableBudget['Spending Ranges (Per Student)'] = pd.cut(SchoolTableBudget['Per Student Budget'], spending_bins, labels = group_names)
# SchoolTableBudget.head()
```

```python
srMath = SchoolTableBudget.groupby('Spending Ranges (Per Student)')[['Average Math Score']].mean()
srMathPct = SchoolTableBudget.groupby('Spending Ranges (Per Student)')[['% Passing Math']].mean()
srRead = SchoolTableBudget.groupby('Spending Ranges (Per Student)')[['Average Reading Score']].mean()
srReadPct = SchoolTableBudget.groupby('Spending Ranges (Per Student)')[['% Passing Reading']].mean()
srOverallPct = SchoolTableBudget.groupby('Spending Ranges (Per Student)')[['Overall Pass %']].mean()
```

```python
srConcat = pd.merge(pd.merge(pd.merge(srMath, srRead, on = 'Spending Ranges (Per Student)'),pd.merge(srMathPct, srReadPct, on='Spending Ranges (Per Student)'),on='Spending Ranges (Per Student)'),srOverallPct, on='Spending Ranges (Per Student)')
srConcat.head()
```
